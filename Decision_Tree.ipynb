```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
df
```

```python
print(df.DESCR)
```

```python
data=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
data
```

```python
data['target']=df.target
```

```python
data
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

## postpruning

```python
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
```

```python
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred_1=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred_1,y_test))

print(classification_report(y_pred_1,y_test))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

### prepruning and Hyperprametertuning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
Parameters={
'criterion':[ 'gini' , 'entropy', 'log_loss'],'splitter':['best', 'random'],
'max_depth':[1,2,3,4,5],
'max_features':['sqrt', 'log2']}
```

```python
from sklearn.model_selection import GridSearchCV
classifier=DecisionTreeClassifier()
clf=GridSearchCV(classifier,param_grid=Parameters)
```

```python
clf
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2,max_features='log2',splitter='random')
```

```python
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_predict,y_test))

print(classification_report(y_predict,y_test))
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))

print(classification_report(y_pred,y_test))
```
